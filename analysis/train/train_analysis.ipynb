```python
import re
import matplotlib.pyplot as plt
from collections import defaultdict
```

### 学习率

```python
info_list = [('e2','lr=1e-2'),('e3','lr=1e-3'),('e4','lr=1e-4')]
cate_list = ['loss','Recall','NDCG','HitRate']

data = defaultdict(dict)

for log,_ in info_list:  
    with open('./'+log, encoding='utf-8') as f:
        lines=f.readlines()
     
    data[log] = defaultdict(list)
        
    for txt in lines:
        if txt[:6] == 'iter: ':  
            info = re.findall('\d+[.\d]*',txt)[1:]
            for i,cate in enumerate(cate_list):
                data[log][cate].append(float(info[i]))

for cate in cate_list:
            
    plt.figure(figsize=(10, 10))

    for log,label in info_list:
        y = data[log][cate]       
        x = [i*1000 for i in range(1,len(y)+1)]
        plt.plot(x,y,label=label)
    plt.legend(loc="lower right",fontsize='xx-large')
    plt.xlabel("batch",fontsize = 25) 
    plt.ylabel(cate,fontsize = 25)
    plt.tick_params(labelsize=15)

    plt.savefig('lr_{}.jpg'.format(cate), bbox_inches='tight')
    #plt.show()
```
